### Code crime Primary Type as an indication of whether it involves a Crime Against Person, Crime Against Property, or Crime Against Society.
> Definitions of what is considered a Crime Against Person, Crime Against Property, or Crime Against Society is according to the Uniform Crime Reporting National Incident-Based Reporting System (NIBRS) produced by the U.S. Department of Justice Federal Bureau of Investigation Criminal Justice Information Services Division.

> In the Summary reporting system, there are two scoring categories: Crimes Against Persons (e.g., Aggravated Assault, Murder, and Forcible Rape) where one offense is counted for each victim and Crimes Against Property (e.g., Burglary, Larceny-theft, and Robbery) where one offense is counted for each distinct operation (except Motor Vehicle Theft where one offense is counted for each stolen vehicle). 
In NIBRS, it was necessary to add a new scoring category to account for crimes such as Drug/Narcotic Offenses, Gambling Offenses, Pornography/Obscene Material, and Prostitution Offenses. They are not Crimes Against Persons as they do not actually involve a victim, nor are they Crimes Against Property because property is not the real object of the crimes. Since these crimes represent society’s prohibitions on engaging in certain types of activity, a new scoring category of Crimes Against Society was created.

>More information on NIBRS can be found at: https://ucr.fbi.gov/nibrs-overview

```python
# load 2010 population dataset
population = pd.read_csv("data/2010 Census Community Area Population.csv")

# load transformed datasets
alleyLightGroup = pd.read_csv("data/Tableau_alleyLightGroup.csv")
buildingGroup = pd.read_csv("data/Tableau_buildingGroup.csv")
graffitiGroup = pd.read_csv("data/Tableau_graffitiGroup.csv")
potholeGroup = pd.read_csv("data/Tableau_potholeGroup.csv")
streetLightGroup = pd.read_csv("data/Tableau_streetLightGroup.csv")
vehicleGroup = pd.read_csv("data/Tableau_vehicleGroup.csv")
crimeGroup = pd.read_csv("data/Tableau_crimeGroup.csv")
populationGroup = pd.read_csv("data/Tableau_populationGroup.csv")

# join tables
geographicDescriptors = socioeconomic.merge(alleyLightGroup, on='Community Area')
geographicDescriptors = geographicDescriptors.merge(buildingGroup, on='Community Area')
geographicDescriptors = geographicDescriptors.merge(graffitiGroup, on='Community Area')
geographicDescriptors = geographicDescriptors.merge(potholeGroup, on='Community Area')
geographicDescriptors = geographicDescriptors.merge(streetLightGroup, on='Community Area')
geographicDescriptors = geographicDescriptors.merge(vehicleGroup, on='Community Area')
geographicDescriptors = geographicDescriptors.merge(crimeGroup, on='Community Area')
geographicDescriptors = geographicDescriptors.merge(population, on='Community Area')

# review merged tables
geographicDescriptors
```

```python
# rename columns for easier modeling & analysis
geographicDescriptors = geographicDescriptors.rename(columns={'2010 Census Population': 'Pop'})
geographicDescriptors = geographicDescriptors.rename(columns={'Per Capita Income': 'Income'})
geographicDescriptors = geographicDescriptors.rename(columns={'Percent of Households With Income Below Poverty Level': 'Poverty'})
geographicDescriptors = geographicDescriptors.rename(columns={'Percent of Crowded Housing': 'CrowdedHousing'})
geographicDescriptors = geographicDescriptors.rename(columns={'Unemployment Rate for Population Age 16 and Over': 'Unemployed'})
geographicDescriptors = geographicDescriptors.rename(columns={'Percent Aged 25 and Over with no High School Diploma': 'NoDiploma'})
geographicDescriptors = geographicDescriptors.rename(columns={'Percent of Population Under Age 18 and Over Age 64': 'Dependency'})
geographicDescriptors = geographicDescriptors.rename(columns={'Hardship Index Score': 'Hardship'})
geographicDescriptors = geographicDescriptors.rename(columns={'Vacant Buildings': 'VacantBuildings'})
geographicDescriptors = geographicDescriptors.rename(columns={'Abandoned Vehicles': 'AbandonedVehicles'})
geographicDescriptors = geographicDescriptors.rename(columns={'Street Lights Out': 'StreetLights'})
geographicDescriptors = geographicDescriptors.rename(columns={'Alley Lights Out': 'AlleyLights'})
geographicDescriptors = geographicDescriptors.rename(columns={'Crime Type_PERSONS': 'CrimePerson'})
geographicDescriptors = geographicDescriptors.rename(columns={'Crime Type_PROPERTY': 'CrimeProperty'})
geographicDescriptors = geographicDescriptors.rename(columns={'Crime Type_SOCIETY': 'CrimeSociety'})
geographicDescriptors = geographicDescriptors.rename(columns={'Total Crime': 'CrimeTotal'})
geographicDescriptors.head()
```

```python
# convert to proper decimal layout to represent percentage
geographicDescriptors['CrowdedHousing'] = (geographicDescriptors['CrowdedHousing'] / 100)
geographicDescriptors['Poverty'] = (geographicDescriptors['Poverty'] / 100)
geographicDescriptors['Unemployed'] = (geographicDescriptors['Unemployed'] / 100)
geographicDescriptors['NoDiploma'] = (geographicDescriptors['NoDiploma'] / 100)
geographicDescriptors['Dependency'] = (geographicDescriptors['Dependency'] / 100)

# convert raw values to percentage of population / per capita
geographicDescriptors['VacantBuildings'] = geographicDescriptors['VacantBuildings'] / geographicDescriptors['Pop']
geographicDescriptors['AbandonedVehicles'] = geographicDescriptors['AbandonedVehicles'] / geographicDescriptors['Pop']
geographicDescriptors['StreetLights'] = geographicDescriptors['StreetLights'] / geographicDescriptors['Pop']
geographicDescriptors['AlleyLights'] = geographicDescriptors['AlleyLights'] / geographicDescriptors['Pop']
geographicDescriptors['Potholes'] = geographicDescriptors['Potholes'] / geographicDescriptors['Pop']
geographicDescriptors['Graffiti'] = geographicDescriptors['Graffiti'] / geographicDescriptors['Pop']
geographicDescriptors['CrimePerson'] = geographicDescriptors['CrimePerson'] / geographicDescriptors['Pop']
geographicDescriptors['CrimeProperty'] = geographicDescriptors['CrimeProperty'] / geographicDescriptors['Pop']
geographicDescriptors['CrimeSociety'] = geographicDescriptors['CrimeSociety'] / geographicDescriptors['Pop']
geographicDescriptors['CrimeTotal'] = geographicDescriptors['CrimeTotal'] / geographicDescriptors['Pop']

geographicDescriptors
```
